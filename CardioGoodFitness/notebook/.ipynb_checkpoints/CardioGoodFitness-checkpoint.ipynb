```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.preprocessing import (
    StandardScaler,
    OneHotEncoder,
    OrdinalEncoder,
    MinMaxScaler
)
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans,AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist
```

```python
data_file = Path('../../data/CardioGoodFitness.csv')
```

```python
df_raw = pd.read_csv(data_file)
```

```python
# Queremos agrupar os clientes em grupos
df_raw.info()
```

```python
df_raw.describe()
```

```python
# temos 3 colunas categóricas: 
# Product passaremos OneHotEncoding
# Gender e MaritalStatus passaremos OrdinalEncodingabs
# vamos normalizar os dados também com StandarScaler
onehot = Pipeline([('onehot', OneHotEncoder(handle_unknown='ignore'))])
ordinal = Pipeline([('ordinal', OrdinalEncoder())])
std = Pipeline([('std', StandardScaler())])
minmax = Pipeline([('minmax', MinMaxScaler())])
```

```python
df_raw
```

```python
all_cols = df_raw.columns
column_preprocess = ColumnTransformer([
    ('pass_ordinal', ordinal, ['Product']),
    ('pass_onehot', onehot, ['Gender', 'MaritalStatus']),
    ('pass_std', std, ['Miles', 'Age', 'Income']),
    ('pass_minmax', minmax, ['Education', 'Usage', 'Fitness'])
    
])

transformer_pipe = Pipeline(steps=[
    ("preprocess", column_preprocess)
])
```

```python
transformer_pipe
```

```python
df = transformer_pipe.fit_transform(df_raw)
```

```python
df = pd.DataFrame(data=df, columns=transformer_pipe.get_feature_names_out(df_raw.columns))
```

```python
k_means = KMeans(n_clusters=3)
```

```python
y_pred = k_means.fit_predict(df)
```

```python
df['result_cluster'] = y_pred
```

```python
df.result_cluster.unique()
```

```python
df.columns
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df.pass_minmax__Education,y=df.pass_std__Miles, c=df.result_cluster)

plt.show()
```

```python
# Creating 20 random initial centroids
kmeans = KMeans(n_clusters=20, init='random', n_init=1, max_iter=1, tol=0.0001, 
                 verbose=0, random_state=None, algorithm='auto')

# Fitting the model
kmeans.fit(df)

# Retreiving cluster centers
init_centroids = kmeans.cluster_centers_
# From 2 to 20 clusters 
k_values = range(2,21)

# Looping over each cluster
final_results=pd.DataFrame()
for k in k_values:
    # Assigning number of clusters
    kmeans = KMeans(n_clusters=k, init=init_centroids[0:k,:], n_init=1, max_iter=10, tol=0.0001, 
                verbose=0, random_state=42, copy_x=True, 
                algorithm='auto')
    # Fitting and predicting labels
    cluster_labels = kmeans.fit_predict(df)
    # Calculating mean intra-cluster distance
    score_wss = round(kmeans.inertia_ ,2)
    # Calculating silhouette score (b-a)/max(a,b)
    silhouette = round(silhouette_score(df, cluster_labels),4)
    # Creating dictionary to store values
    aux_results = {'score_wss': score_wss, 'silhouette': silhouette}
    # Assigning dictionary to dataframe
    aux_results = pd.DataFrame(aux_results, index = [k])
    # Concatenating aux values with final results
    final_results = pd.concat([final_results, aux_results], axis =0)
```

```python
# Plotting results
fig, ax = plt.subplots(2,1,figsize=(12,8))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.35)

ax[0].plot(final_results.index,final_results.score_wss)
ax[0].set_xlabel('Number of Clusters')
ax[0].set_ylabel('Within Cluster Sum of Squares')
ax[0].set_title('Elbow')
ax[0].set_xticks(k_values)
ax[0].set_xticklabels(k_values)
ax[0].grid()


ax[1].plot(final_results.index,final_results.silhouette)
ax[1].set_xlabel('Number of Clusters')
ax[1].set_ylabel('Silhouette Score')
ax[1].set_title('Silhouette')
ax[1].set_xticks(k_values)
ax[1].set_xticklabels(k_values)
ax[1].grid()

fig.tight_layout()
plt.show()
```

```python
af_type = 'euclidean'
lk_type = 'complete'
dist_mat = pdist(df, metric=af_type)

# Creating hierarchy
linkage_matrix = hierarchy.linkage(dist_mat, method=lk_type, metric=af_type)

fig, ax = plt.subplots(1,1,figsize=(16, 8))

hierarchy.dendrogram(linkage_matrix,
                     color_threshold=0,
                     ax=ax)
ax.set_title('Hierarchical Clustering Dendrogram ('+lk_type+')')
ax.set_xlabel('sample index')
ax.set_ylabel('distance')
ax.set_xticks(linkage_matrix[::3,3])
fig.tight_layout()
plt.show()
```
