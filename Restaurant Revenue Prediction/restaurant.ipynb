```python
# o conjunto de treino é muito pequeno provavelmente uma LSTM n irá convergir
# vamos passar um SVR
df_train['Open Date'] = pd.to_datetime(df_train['Open Date'])
```

```python
df_train = df_train.sort_values(by='Open Date')
```

```python
# por ter uma ordem temporal vamos usar TimesSeriesSplit como validação cruzada
X = df_train.drop(['revenue', 'Id'], axis=1)
y = df_train.revenue
```

```python
X.iloc[100:]
```

```python
X_train, X_val = X.iloc[:100], X.iloc[100:]
```

```python
y_train, y_val = y.iloc[:100], y.iloc[100:]
```

# Pré processando os dados

```python
depara_cidades = df_train.City.value_counts().to_dict()
```

```python
def depara_cidades_func(
    df=X_train,
    coluna='City',
    mapping_dict=depara_cidades
):
    for k, v in mapping_dict.items():
        df[coluna] = df.loc[:, coluna].replace(k,v)
    return df
```

```python
depara_cidades_func()
```

```python
depara_cidades_func(X_train)
```

```python
df_train.head()
```

```python
colunas_one = ['Type']
colunas_ordinal = ['City Group']
colunas_numericas = X_train.columns.to_list()[4:]
```

```python
df_train['City Group'].value_countss()
```

```python
# Fazendo os pipes
onehot = Pipeline([('one', OneHotEncoder(handle_unknown='ignore'))])
ordinal = Pipeline([('ord', OrdinalEncoder())])
num = Pipeline([('normalization', StandardScaler())])
```

```python
prep = ColumnTransformer([
    ('one_ct', onehot, colunas_one),
    ('oridnal_ct', ordinal, colunas_ordinal),
    ('num_ct', num, colunas_numericas)
])
```

```python
prep
```

```python
model = Pipeline([
    ('preprocess', prep),
    ('model', SVR())
])
```

```python
model.fit(X_train, y_train)
```

```python
ttr = TransformedTargetRegressor(regressor=model, func=np.log1p, inverse_func=np.expm1)
```

```python
ttr
```

```python
cv = TimeSeriesSplit(n_splits=3)
```

```python
from sklearn.metrics import get_scorer_names 
```

```python
get_scorer_names()
```

```python
validation = cross_validate(
    estimator=ttr,
    X=X_train,
    y=y_train,
    cv=cv,
    return_train_score=True,
    scoring='neg_mean_absolute_error'
)
```

```python
val_table = pd.DataFrame(validation)
```
