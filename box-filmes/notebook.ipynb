```python
# Pegue o genero da coluna genres
def get_genre(value: list):
    if not value:
        return ''
    value = ast.literal_eval(value)
    value = [value[i].get('name') for i in range(len(value))]
    value.sort()
    value = '-'.join(value)
    return value

# Faça um regex e pege o site homepage
def get_site(value):
    if isinstance(value ,float):
        return ''
    regex_comp = re.compile(r"https?://([A-Za-z_0-9.-]+).*")
    value = regex_comp.search(value).group(1)
    return value

# trata o texto
def trata_overview(texto):
    vectorize = TfidfVectorizer(stop_words='english')
    texto = vectorize.fit_transform(texto)
    
    svd = TruncatedSVD(n_components=10, n_iter=10, random_state=42)
    texto_red = svd.fit_transform(texto)
    names = [f'overview_{i}' for i in range(10)]
    data = pd.DataFrame(texto_red, columns=names)
    
    return data

# exlcui a coluna overview e coluca as novas criadas
def put_overview_svd(data, data_text):
    data = data.drop(['overview'], axis=1)
    data = pd.concat([data, data_text], axis=1)
    return data

def years(data):
    current_year = date.today().year
    data = data.year
    return current_year - data

def get_produtora(value):
    if isinstance(value ,float):
        return ''
    value = ast.literal_eval(value)
    value = [value[i].get('name') for i in range(len(value))]
    return value[0]
```

```python
dict_func = {
    get_genre: 'genres',
    get_site: 'homepage',
    years: 'release_date',
    get_produtora: 'production_companies'
}
```

```python
for k, v in dict_func.items():
    df_train[v] = df_train[v].apply(lambda x: k(x))
```

```python
series_text = df_train.overview
df_text = trata_overview(series_text)
```

```python
######################## CATEGÓRICAS #####################################
# genres, status
ordinal = Pipeline(steps=[
    ('simple_ordinal', SimpleImputer(missing_values='', strategy='constant', fill_value='N/A')),
    ('ord_enc', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

# homepage, original_language
onehot = Pipeline(steps=[
    ('simple_onehot', SimpleImputer(missing_values='', strategy='most_frequent')),
    ('one_enc', OneHotEncoder(handle_unknown='ignore', min_frequency=3))
])

######################## NÚMERICAS #####################################
# budget, popularity, release_date, runtime
num = Pipeline(steps=[
    ('simple', SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=-1)),
    ('normalization', MinMaxScaler())
])
```

```python
preprocess = ColumnTransformer([
    ("categoricas_ordinal", ordinal, ['genres', 'status']),
    ("categoricas_onehot", onehot, ['homepage', 'original_language', 'production_companies']),
    ("numericas", num, ['budget', 'popularity', 'release_date', 'runtime'])
], remainder='passthrough')
```

```python
model = Pipeline([
    ('preprocess', preprocess),
    ('model', AdaBoostRegressor())
])
```

```python
model
```

# Target

```python
target_pipeline = TransformedTargetRegressor(
    regressor=model,
    transformer=PowerTransformer(method='box-cox')
)
target_pipeline
```

```python
# errors por bins
def rmse_bins(y_test, y_pred, intervalo=10000):
    y_pred_final = pd.Series(y_pred)
    y_test_final = y_test.reset_index()
    df = pd.concat([y_pred_final, y_test_final], axis=1)
    df.rename({'revenue': 'y_test', 0: 'y_pred'}, axis=1, inplace=True)
    df['bin'] = (df['y_test'] // intervalo).abs()
    df['intervalo_min'] = df['bin'] * intervalo
    df['intervalo_max'] = (df['bin'] + 1) * intervalo
    print(df.groupby(['bin', 'intervalo_min', 'intervalo_max']).apply(
        lambda x: np.sqrt(mean_squared_error(x['y_test'], x['y_pred']))))
    return df
```
