```python
modelos = [
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    LogisticRegression(),
    SVC(),
    DecisionTreeClassifier(max_depth=5),
    KNeighborsClassifier(),
    MLPClassifier()
]

def fit_models():
    models = {}
    for i, base in enumerate(dfs_train):
        X = base.drop('TARGET', axis=1)
        y = base.TARGET
        X_train, X_val, y_train, y_val = train_test_split(
            X,
            y,
            stratify=y,
            test_size=0.05
        )
        base_estimator=choice(modelos)
        modelo = BaggingClassifier(
            base_estimator=base_estimator,
            n_estimators=300,
            max_features=30,
            max_samples=0.9,
            bootstrap_features=True,
            oob_score=True
        )
        modelo.fit(X_train, y_train)
        y_pred = modelo.predict(X_val)
        print(f'modelo: {modelo}')
        print(classification_report(y_val, y_pred))
        print()
        models[f'{modelo}_{i}'] = modelo
    return models
```

```python
# por fim agora variamos o threshold 
def select_threshold(
    y_true: np.array,
    y_pred: np.array,
    metric: callable,
    list_threshold: float
) -> float:
    
    val = []
    for threshold in list_threshold:
        y_p = np.where(y_pred >= threshold, 1, 0)
        v = f1_score(y_true, y_p)
        val.append((threshold, v))
    
    sorted_threshold = sorted(val, key=lambda p: p[1])
    
    return sorted_threshold
```

```python
# por fim agora variamos o threshold 
def select_threshold(
    y_true: np.array,
    y_pred: np.array,
    metric: callable,
    list_threshold: float
) -> float:
    
    val = []
    for threshold in list_threshold:
        y_p = np.where(y_pred >= threshold, 1, 0)
        acc = accuracy_score(y_true, y_p)
        rec = recall_score(y_true, y_p)
        prec = precision_score(y_true, y_p)
        f1 = f1_score(y_true, y_p)
        
        val.append((threshold, acc, rec, prec, f1))
    
    valor = np.array(val)
    
    return valor
```

```python
sorted_threshold = select_threshold(
    y_true = y_test_val,
    y_pred = results_mean,
    metric = accuracy_score,
    list_threshold = threshold
)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1,ncols=4)
ax1.plot(sorted_threshold[:,0], sorted_threshold[:,1], color='r')
ax1.set_title('acc')

ax2.plot(sorted_threshold[:,0], sorted_threshold[:,2], color='r')
ax2.set_title('rec')

ax3.plot(sorted_threshold[:,0], sorted_threshold[:,3], color='r')
ax3.set_title('prec')

ax4.plot(sorted_threshold[:,0], sorted_threshold[:,4], color='r')
ax4.set_title('f1-score')
ax4.set_xlabel('limiar')

plt.tight_layout()
plt.show()
```
